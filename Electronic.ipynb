```python
# info function
print("\nInfo function:")
print(data.info())

# Describe function
print("\nDescribe function:")
print(data.describe())

# Head function
print("\nHead function:")
print(data.head())

# Tail function
print("\nTail function")
print(data.tail())

# duplicate function
print("\nDuplicate function:")
print(data.duplicated())
```

```python
# Total number of Purchase
Total_Purchase=(data["purchase_made"]=="Yes").sum()
print(f"Total number of purchases: {Total_Purchase}")

# Total phage view
Total_View = data["pages_visited"].sum()
print(f"Total number of phage visted: {Total_View}")

# Product category and items added to cart
view_item = data.groupby("product_category")["items_added_to_cart"].sum()
print("\nProduct category and items added to cart:")
print(view_item)

# device type and phage
avg = data.groupby("device_type")["pages_visited"].agg(["mean","sum"])
print("\ndevice type and phage visted:")
print(avg)

# Product sale and it's category
sale = data.groupby("product_category")["purchase_made"].count()
print("\nProduct sale and it's category:")
print(sale)

# Toal sale
TP = data["order_value"].sum()
print("\nTotal Revenue")
print(f"Total Purchase: {TP}")
```

```python
sale_chennal = data.groupby("marketing_channel")["purchase_made"].count()
print(sale_chennal)
```

```python
plt.figure(figsize=(10, 6))
sale.plot(kind="bar", color="lightseagreen")
plt.title(" Product sale and it's category")
plt.xlabel("Product Category")
plt.ylabel("Number of Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sale_chennal.plot(kind="bar",color="Lightseagreen")
plt.title("Product sale through sites")
plt.xlabel("Marketing Channel")
plt.ylabel("Number of Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
customer_data = data.groupby('user_id').agg({
    'session_duration': 'sum',
    'pages_visited': 'sum',
    'items_added_to_cart': 'sum',
    'order_value': 'sum',
    'purchase_made': lambda x: (x == 'Yes').sum()  # Count purchases
}).reset_index()

# Create customer segments
customer_data['customer_type'] = 'Window Shopper'
customer_data.loc[customer_data['order_value'] > 0, 'customer_type'] = 'One-Time Buyer'
customer_data.loc[customer_data['order_value'] > 1000, 'customer_type'] = 'VIP Customer'
customer_data.loc[customer_data['purchase_made'] > 1, 'customer_type'] = 'Repeat Customer'

print("Customer Segmentation:")
print(customer_data['customer_type'].value_counts())
```

```python
customer_data.plot(kind="bar",color="lightseagreen")
plt.title("Customer Segmentation")
plt.xlabel("Customer Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
customer_type_counts = customer_data['customer_type'].value_counts()

plt.figure(figsize=(8, 8))
customer_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90,color='#FF9999')
plt.title('Customer Segmentation')
plt.ylabel('') # Remove the default 'None' label
plt.show()
```
