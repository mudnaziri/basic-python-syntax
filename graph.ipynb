```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [120, 150, 180, 90, 210],
    'Expenses': [80, 90, 100, 110, 95]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.plot(x='Month', y=['Sales', 'Expenses'],
        kind='line',
        title='Monthly Sales vs Expenses',
        figsize=(10,6))
plt.ylabel('Amount ($)')
plt.show()
```

```python
df.plot(x='Month', y="Sales",
       kind='bar',
       color='skyblue',
       title='Monthly Sales')
plt.ylabel('Sales ($)')
plt.show()
```

```python
df.plot(x="Month", y= "Expenses",
        kind="barh",
        color="blue",
        title="Monthly Expenses")
plt.ylabel("expenses($)")
plt.show()
```

```python
df.plot(y='Sales', x='Month',
        kind='pie',
        autopct='%1.1f%%',
        title='Sales Distribution')
plt.ylabel('')  # Remove y-label
plt.show()
```
