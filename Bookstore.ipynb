```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Date': ['2023-10-27', '2023-10-27', '2023-10-28', '2023-10-28',
             '2023-10-28', '2023-10-29', '2023-10-29', '2023-10-29'],
    'Book_Title': ['The Midnight Library', 'A Brief History...', 'The Silent Patient',
                   'Where the Crawdads..', 'Sapiens', 'The Guest List',
                   'Educated', 'The Midnight Library'],
    'Genre': ['Fiction', 'Non-Fiction', 'Mystery', 'Fiction',
              'Non-Fiction', 'Mystery', 'Non-Fiction', 'Fiction'],
    'Price': [16.99, 22.50, 14.50, 18.75, 20.00, 13.99, 17.00, 16.99],
    'Quantity_Sold': [3, 1, 2, 4, 1, 2, 1, 1],
    'Customer_Age': [34, 52, 29, 41, 35, 27, 58, 23]
})

print(data)
```

```python
print("\nInfo function")
print(data.info())

#describe
print("\ndescribe function")
print(data.describe())

```

```python
# Revenue Count
data["Total_Revaune"] = data["Price"]*data["Quantity_Sold"]
print(data)
```

```python
# Quantity of Book were sold
data["Quantity_Sold"].sum()
```

```python
# Count of Fiction book
Fiction_Count = (data["Genre"]=="Fiction").sum()
print(f"Fiction Book Count: {Fiction_Count}")

# Count Non-Fiction books
non_fiction_count = (data["Genre"] == "Non-Fiction").sum()
print(f"Number of Non-Fiction books: {non_fiction_count}")

# Count Mystery books
mystery_count = (data["Genre"] == "Mystery").sum()
print(f"Number of Mystery books: {mystery_count}")
```

```python
data["Young_Adults"] = (data["Customer_Age"] >= 18) & (data["Customer_Age"] <= 25)
data["Adults"] = (data["Customer_Age"] >= 26) & (data["Customer_Age"] <= 35)
data["Senior_Adults"] = (data["Customer_Age"] >= 36) & (data["Customer_Age"] <= 60)

print(data)
```

```python
YA_count=(data["Young_Adults"]==True).sum()
print(f"Young Adults count: {YA_count}")

A_count=(data["Adults"]==True).sum()
print(f"Adults count: {A_count}")

SA_count=(data["Senior_Adults"]==True).sum()
print(f"Senior Adults count: {SA_count}")
```

```python
Book_count = (data["Quantity_Sold"]==data["Young_Adults"]).sum()
print(f"Book count: {Book_count}")

book_count = (data["Quantity_Sold"]==data["Adults"]).sum()
print(f"Book count: {book_count}")

book_count1 = (data["Quantity_Sold"]==data["Senior_Adults"]).sum()
print(f"Book count: {book_count1}")
```

```python
b = data.groupby("Genre")["Quantity_Sold"].sum()
print(b)
```

```python
plt.figure(figsize=(10, 6))
book_counts = data['Book_Title'].value_counts()
book_counts.plot(kind='bar', color='lightseagreen')
plt.title('Most Popular Books Among Customers Aged 25-60')
plt.xlabel('Book Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.ylabel('Number of Copies Sold')
plt.show()
```
